```python


# Defining node
class Node:
    def __init__(self, value=None,
                 next_element=None):
        self.val = value
        self.next = next_element

# Implementing stack class


class stack:

    def __init__(self):
        self.head = None
        self.length = 0

    # Adding a node to the stack
    def insert(self, data):
        self.head = Node(data, self.head)
        self.length += 1

    # Return the top position of the stack
    def pop(self):
        if self.length == 0:
            return None
        else:
            returned = self.head.val
            self.head = self.head.next
            self.length -= 1
            return returned

    # Check if stack is empty or not and return true is not empty
    def not_empty(self):
        return bool(self.length)

    # Return the top position of the stack
    def top(self):
        return self.head.val

# Generate random maze


def generate_maze(r, c, Initial, Final):
    row, column = r, c

    # Initializing the matrix with only blockers
    status = list(list('*' for _ in range(column))
                  for _ in range(row))

    # Auxiliary matrix to avoid cycles when making the path
    explored = list(list(False for _ in range(column))
                    for _ in range(row))
    parent = list(list((-1, -1)
                       for _ in range(column)) for _ in range(row))

    to_visit = stack()

    # Insert the start point to the stack
    to_visit.insert(Initial)

    # Exploring all the nodes until all the generated one's are visited
    while to_visit.not_empty():

        # When exploring a particular node, remove it from the stack and add it to the explored matrix
        x, y = to_visit.pop()
        explored[x][y] = True

        # For any valid neighbor positions of the node, we ensure that it is not already in the path and is not explored.
        if (x + 1 < row) and status[x + 1][y] == 1 \
                and parent[x][y] != (x + 1, y):
            continue
        if (0 < x) and status[x-1][y] == 1 \
                and parent[x][y] != (x-1, y):
            continue
        if (y + 1 < column) and status[x][y + 1] == 1 \
                and parent[x][y] != (x, y + 1):
            continue
        if (y > 0) and status[x][y-1] == 1 \
                and parent[x][y] != (x, y-1):
            continue

        # Removing the blocker and making it a part of path
        status[x][y] = 1

        # Array to shuffle neighbours before
        # insertion
        shuffle_neighbours = []

        # To ensure that the row and column index is within the maze size and it is not explored
        if (x + 1 < row) and explored[x + 1][y] == False:

            # Change the neighbor value to explored
            explored[x + 1][y] = True

            # Adding this neighbor position to the stack so that we can re-visit again
            shuffle_neighbours.append((x + 1, y))

            # Saving the parent position of this neighbor node to the parent matrix
            parent[x + 1][y] = (x, y)

        if (0 < x) and explored[x-1][y] == False:

            # Change the neighbor value to explored
            explored[x-1][y] = True

            # Adding this neighbor position to the stack so that we can re-visit again
            shuffle_neighbours.append((x-1, y))

            # Saving the parent position of this neighbor node to the parent matrix
            parent[x-1][y] = (x, y)

        if (y + 1 < column) and explored[x][y + 1] == False:

            # Change the neighbor value to explored
            explored[x][y + 1] = True

            # Adding this neighbor position to the stack so that we can re-visit again
            shuffle_neighbours.append((x, y + 1))

            # Saving the parent position of this neighbor node to the parent matrix
            parent[x][y + 1] = (x, y)

        if (y > 0) and explored[x][y-1] == False:

            # Change the neighbor value to explored
            explored[x][y-1] = True

            # Adding this neighbor position to the stack so that we can re-visit again
            shuffle_neighbours.append((x, y-1))

            # Saving the parent position of this neighbor node to the parent matrix
            parent[x][y-1] = (x, y)

        # Verifies if the target is in the list of neighbors for the current node and replacing it with the first element if found.
        check = False
        while len(shuffle_neighbours):

            # Pick random node to be explored from the list of nodes to be visited from the stack
            neighbour = shuffle_neighbours.pop(
                randint(0, len(shuffle_neighbours)-1))

            # If the random nose picked is the target then save it
            if neighbour == Final:
                check = True

            # add target as the first element of the stack
            else:
                to_visit.insert(neighbour)

        # Adds the target to the top of the stack
        if check:
            to_visit.insert(Final)

    # Initialising the start and target positions
    xi, yi = Initial
    xf, yf = Final
    status[xi][yi] = 'Source'
    status[xf][yf] = 'Target'

    return status
```

```python

def drawGrid(a):

    for row in range(len(a)):
        for column in range(len(a[0])):
            color = WHITE
            if a[row][column] == '*':
                color = RED
            if a[row][column] == 'S':
                color = BLUE
            if a[row][column] == 'T':
                color = YELLOW
            pygame.draw.rect(SCREEN,
                             color,
                             [(MARGIN + WIDTH) * column + MARGIN,
                              (MARGIN + HEIGHT) * row + MARGIN,
                              WIDTH,
                              HEIGHT])
```

```python
BLACK = (0, 0, 0)
WHITE = (200, 200, 200)
GREEN = (0, 255, 0,)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
YELLOW = (255, 255, 0)
WINDOW_HEIGHT = 850
WINDOW_WIDTH = 850
WIDTH = 6.3
HEIGHT = WIDTH
MARGIN = 2
FPS = 30

SCREEN = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Python Maze Generator")
clock = pygame.time.Clock()
SCREEN.fill(BLACK)
```
